```python
##RAG
from langchain.chains import RetrievalQA
from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.llms import BaseLLM
from langchain.schema import LLMResult, Generation
# from langchain.chains import create_retrieval_chain
# from langchain.chains.combine_documents import create_stuff_documents_chain
# from langchain.chains.question_answering import load_qa_chain

# Initialize the model and tokenizer
model_name = "facebook/opt-6.7b"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# Define the custom LLM interface with LangChain
class OPTLLM(BaseLLM):
    def _generate(self, prompt: str, stop: list = None, **kwargs) -> str:
        # Generate the response using the OPT-6.7B model
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

        # The `stop` parameter can be used to control the stopping criteria for generation
        if stop:
            stop_tokens = tokenizer(stop, return_tensors="pt").input_ids
        else:
            stop_tokens = None

        # Generate the response from the model
        outputs = model.generate(**inputs, max_new_tokens=100, stopping_criteria=stop_tokens, **kwargs)

        # Decode the generated tokens and create a response
        response = tokenizer.decode(outputs[0], skip_special_tokens=True)
        print(f"Response: {response}")

        return LLMResult(
            generations=[[Generation(text=response)]],
            llm_output={"generated_tokens": outputs.tolist()}  # ✅ Convert tensor to list (dict format)
        )

    def _llm_type(self) -> str:
        return "custom"

# Instantiate the LLM with the OPT-6.7B model
optllm = OPTLLM()

retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
# qa_chain = load_qa_chain(llm, chain_type="stuff")
rag_chain = RetrievalQA.from_llm(llm=optllm, retriever=retriever)

from functools import lru_cache

# Define a simple cache function
@lru_cache(maxsize=100)  # Store last 100 responses
def get_cached_rag_response(query):
    return rag_chain.run(query)


print("Initiated rag chain")
# Example query
query = 'Show me a red shirt for men'
response = rag_chain.get_cached_rag_response(query)
print(response)
```

```python
import ray
from ray import tune
from ray.tune.schedulers import ASHAScheduler
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader
import torch
from sentence_transformers.losses import MultipleNegativesRankingLoss
import pandas as pd
from ray.air import RunConfig
from ray.tune import Callback
ray.shutdown()
ray.init(log_to_driver=False)

# from transformers import TrainingArguments

# training_args = TrainingArguments(
#     output_dir="./results",
#     logging_dir="./logs",
#     report_to="none",  # ✅ Prevents wandb from being used
# )

import os
# os.environ["WANDB_DISABLED"] = "true"
# os.environ["WANDB_MODE"] = "offline"
# os.environ["WANDB_SILENT"] = "true"
# os.environ["WANDB_API_KEY"] = " "  # Clear API key
# os.environ["WANDB__SERVICE_WAIT"] = "0"
# os.environ["WANDB_DIR"] = "/dev/null"

# import sys
# if "wandb" in sys.modules:
#     del sys.modules["wandb"]

df_tune_sample = df_master.sample(500)
train_data = []

for _, row in df_tune_sample.iterrows():
    # Select similar products from the same category
    similar_products = df_master[df_master['product_type'] == row['product_type']]

    # Ensure there are valid similar products
    if not similar_products.empty:
        sampled_product = similar_products.sample(1).iloc[0]  # Randomly pick one similar product
        train_data.append(InputExample(texts=[row['enhanced_product_desc'], sampled_product['enhanced_product_desc']]))

print(f"Total Training Samples: {len(train_data)}")  # Debugging

# Define fine-tuning function
def train_embedding_model(config):
    model = SentenceTransformer(config["model_name"])  # Load base model
    train_dataloader = DataLoader(train_data, batch_size=config["batch_size"], shuffle=True, collate_fn=model.smart_batching_collate)
    train_loss = MultipleNegativesRankingLoss(model)

    model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=config["epochs"], warmup_steps=100, optimizer_params={'lr': config["lr"]})

    # Report loss for hyperparameter tuning
    # tune.report(loss=train_loss(model, next(iter(train_dataloader))[0]))

    # Save the fine-tuned model
    model.save("fine_tuned_embedding_model")

# Run Ray Tune hyperparameter search
search_space = {
    "lr": tune.loguniform(1e-5, 1e-3),
    "batch_size": tune.choice([2, 4, 6]),
    "epochs": tune.choice([2, 3, 4]),
    "model_name": tune.choice(['sentence-transformers/all-MiniLM-L6-v2', 'sentence-transformers/multi-qa-mpnet-base-dot-v1'])
}

class DisableWandbCallback(Callback):
    def on_trial_start(self, iteration, trials, trial, **info):
        os.environ["WANDB_DISABLED"] = "true"
        os.environ["WANDB_MODE"] = "offline"
        os.environ["WANDB_SILENT"] = "true"
        os.environ["WANDB_API_KEY"] = " "

    def on_trial_result(self, iteration, trials, trial, result, **info):
        if "wandb" in result:
            del result["wandb"]  # ✅ Delete wandb logs if they appear

    def on_experiment_end(self, trials, **info):
        os.environ["WANDB_DISABLED"] = "true"


tuner = tune.Tuner(
    train_embedding_model,
    param_space=search_space,
    tune_config=tune.TuneConfig(num_samples=10, metric="loss", mode="min"),
    run_config=RunConfig(log_to_file=False, verbose=1, callbacks=[DisableWandbCallback()])
)

tuner.fit()
```

```python
import torch
from sentence_transformers import SentenceTransformer, InputExample, losses
from torch.utils.data import DataLoader
from peft import LoraConfig, get_peft_model
import pandas as pd

# Load the base model
model_name = "sentence-transformers/all-MiniLM-L6-v2"
base_model = SentenceTransformer(model_name)

# Apply LoRA configuration
lora_config = LoraConfig(
    r=8,  # Low-rank dimension
    lora_alpha=16,  # Scaling factor
    target_modules=["attention.self.query", "attention.self.key"],  # Target attention layers
    lora_dropout=0.05,
    bias="none"
)

# Wrap the model with LoRA
peft_model = get_peft_model(base_model, lora_config)

# Load dataset (assuming df_master exists)
df_sample = df_master.sample(5000)
train_data = [
    InputExample(texts=[row['enhanced_product_desc'], df_master[df_master['product_type'] == row['product_type']].sample(1).iloc[0]['enhanced_product_desc']])
    for _, row in df_sample.iterrows()
]

# DataLoader setup
train_dataloader = DataLoader(train_data, batch_size=4, shuffle=True)

# Define loss function
train_loss = losses.MultipleNegativesRankingLoss(peft_model)

# Fine-tune the model
peft_model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=3,
    warmup_steps=100
)

# Save fine-tuned model
peft_model.save("fine_tuned_lora_model")
print("Fine-tuned model saved successfully!")
```
